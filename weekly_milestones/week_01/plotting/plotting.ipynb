```python
import pandas as pd

df = pd.read_csv("../data/gapminder_total_fertility.csv")
df
```

```python
df = df[['Total fertility rate','1950', '1955', '2000', '2007', '2015']]
df
```

```python
df_short= df[( df['Total fertility rate'].isin(['Germany', 'India', 'Bulgaria', 'Kenya']))]
df_short
```

```python
from matplotlib import pyplot as plt
df_short.index = df_short['Total fertility rate']
df_short.columns
```

```python
dft = df_short.drop(columns=['Total fertility rate'])
dft
```

```python
dft.plot.bar(figsize=(15,7))

plt.savefig('bars.png')
```
