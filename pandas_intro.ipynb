```python
import numpy as np
data = {'spices': ['parsley', 'sage', 'rosemary', 'thyme'],
        'value': [1.2, 3.4, np.nan, 5.6],
        'good_for_spaghetti': [False, False, True, True]}

labels = ['a', 'b', 'c', 'd']
df=pd.DataFrame(data,index=labels)
df
```

```python
import pandas

spices = ['One-Hot Chili Peppers',
          'Bayesian Basil',
          'Tensor Thyme'
          'Linear Lavender',
          'Artificial Neural Nutmeg',
          'Polynomial Peppermint',
          'Sigmoid Saffron'
          ]
participants = [2, 6, 9, 9, 9, 8]
data = {'name': spices,
        'participants': participants}

labels = ['a', 'b', 'c', 'd', 'e', 'f']
df=pd.DataFrame(data,index=labels)

df
```

```python

print(df.sort_values(by='participants', ascending=False))

print("\ntotal participants:", df['participants'].sum())
```

```python
cont = pd.read_csv('data/continents.csv', sep = ';')
```

```python
# 2. display the number of rows and columns
cont.shape
cont.columns
```

```python
# 3. display the first 5 countries in the alphabet
cont.sort_values(['country']).head()
```

```python
# 4. which continent has the most countries?
cont.groupby('continent').count()[cont.groupby('continent').count()['country'] == cont.groupby('continent').count()['country'].max()]
```

```python
# 5. find out on which continent Cyprus is
cont[(cont['country'] == "Cyprus")]
```

```python
# 6. define a DataFrame with all African countries
africa = cont[(cont["continent"] == "Africa")]
africa
```

```python

# 7. define a DataFrame with countries 80 through 100
subset =  cont.iloc[80:100]
subset
```

```python
# 8. select every second country
subset2 = cont.iloc[::2]
subset2
```
