```python
life = pd.read_excel('data/gapminder_lifeexpectancy.xlsx', index_col=0)
fert = pd.read_csv('data/gapminder_total_fertility.csv', index_col=0)
```

```python
import matplotlib.pyplot as plt

lifeexp = life[[1950, 1975, 2000, 2015]]

lifeexp[2015].hist()

plt.savefig('plots/histo.png')
```

```python
#lifeexp[2015].hist(bins=5)
#lifeexp[2015].hist(bins=10)
plt.title('Histogram of life expectancies for the 2015 year')
plt.xlabel('value')
plt.ylabel('frequency')
plt.axis([40.0, 90.0, 0.0, 25.0])

lifeexp[2015].hist(bins=25, color='#44ff55', alpha=0.75, histtype='barstacked')
plt.savefig('plots/histo.svg', dpi=150)

```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(19680801)

n_bins = 10
x = lifeexp[[1975, 2000, 2015]]

fig, ((ax0, ax1), (ax2, ax3)) = plt.subplots(nrows=2, ncols=2)

colors = ['red', 'tan', 'lime']
ax0.hist(x, n_bins, density=True, histtype='bar', color=colors, label=colors)
ax0.legend(prop={'size': 10})
ax0.set_title('Life bars with legend')

ax1.hist(x, n_bins, density=True, histtype='bar', stacked=True)
ax1.set_title('Life stacked bar')

ax2.hist(x, n_bins, histtype='step', stacked=True, fill=False)
ax2.set_title('Life stack step (unfilled)')

# Make a multiple-histogram of data-sets with different length.
x_multi = [np.random.randn(n) for n in [10000, 5000, 2000]]
ax3.hist(x_multi, n_bins, histtype='bar')
ax3.set_title('Life different sample sizes')

fig.tight_layout()
plt.show()
```
